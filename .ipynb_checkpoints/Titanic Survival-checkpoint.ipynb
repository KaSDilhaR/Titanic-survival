```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['Name','PassengerId','Parch','Ticket','Cabin','Embarked','Fare'],axis='columns',inplace=True)
df.head()
```

```python
iv=df.drop('Survived',axis='columns')
dv=df.Survived
```

```python
iv
```

```python
dv
```

```python
dummies=pd.get_dummies(iv.Sex,dtype=int)
```

```python
dummies.head(200)
```

```python
iv=pd.concat([iv,dummies],axis='columns')
iv.head()
```

```python
iv.drop(['Sex','male'],axis='columns',inplace=True)
```

```python
iv.head()
```

```python
iv.head(200)
```

```python
iv.columns[iv.isna().any()]
```

```python
import math
median_age=math.floor(df.Age.median())
print(median_age)
```

```python
iv['Age'].fillna(value=median_age,inplace=True)
```

```python
iv.columns[iv.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iv,dv,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
